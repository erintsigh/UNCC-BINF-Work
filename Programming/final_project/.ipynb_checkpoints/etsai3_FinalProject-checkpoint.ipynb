### Overview: 
The focus of this project is on covid19 data in the U.S. and whether or not there are differences between races, here I only look at death data. Is one race more susceptible than others? I use data from apmresearchlab and data from the U.S. census bureau to also look at percentages of race in the U.S. per state. 

Start with importing pandas and R (rpy2). Reading in a csv file from apmresearchlab and from governing/data from US census bureau.
This file contains the deaths of each race from each state that they had information from and I set the index as the location/state. I transposed the data and then added a row with the column totals, called total.

```python
import pandas as pd
%load_ext rpy2.ipython

path = "/Users/erintsigh/python_spring2020/final_project/"

#reading in the csv files as dataframes for covid by races
US_demographic_perstate = pd.read_csv(path + "demographic_of_US.csv").set_index('State').dropna()
allraces_coviddeaths = pd.read_csv(path + "covid_byrace/" + "data-allraces.csv").set_index('State')
#.dropna()
allraces_coviddeaths['Total'] = allraces_coviddeaths.sum(axis=1)
allraces_coviddeaths
```

```python
allraces_coviddeaths.sort_values(by="Total",ascending=False,inplace=True)
allraces_coviddeaths_top = allraces_coviddeaths.head().T
allraces_coviddeaths_top
plt_deathsraces = allraces_coviddeaths_top.plot.bar()
plt_deathsraces
```

Then I combine my two datasets (US census bureau and covid deaths data) because I want to divide the covid deaths per race per state by their actual population data from the US census, so that I can determine if one race actually is more at risk than the others. And drop all the states that have NaN values.

Then I created new columns of the number of covid deaths per race, per state divided by the total number of people per that race, per state, multiplied by 100K (or per capita). I did this to weight the number of deaths per the amount of people per that race for that state because every state has a different percentage for each race. If I didn't do this, it would be saying that every race has an equal population per state. So after I weighted each race, I created a new table with all the races per state per capita, so this table basically says how many people from that particular race will die from covid per 100K people. 

I also added a total average deaths per state in a new column under, 'Total deaths per 100K', and I added the average rate of deaths per capita for each race at the bottom under the row 'Average'.

```python
coviddeaths_uscensus = allraces_coviddeaths.combine_first(US_demographic_perstate).dropna()

coviddeaths_uscensus['White deaths per 100K'] = (coviddeaths_uscensus['# of White Deaths']/coviddeaths_uscensus['White'])*100000
coviddeaths_uscensus['Black deaths per 100K'] = (coviddeaths_uscensus['# of Black Deaths']/coviddeaths_uscensus['Black'])*100000
coviddeaths_uscensus['Hispanic deaths per 100K'] = (coviddeaths_uscensus['# of Hispanic Deaths']/coviddeaths_uscensus['Hispanic'])*100000
coviddeaths_uscensus['Asian deaths per 100K'] = (coviddeaths_uscensus['# of Asian Deaths']/coviddeaths_uscensus['Asian'])*100000
coviddeaths_uscensus['Total deaths per 100K'] = (coviddeaths_uscensus['Total']/coviddeaths_uscensus['Total Population'])*100000
coviddeaths_uscensus
```

Then I created a new dataframe of just the deaths per race per capita and added a row, called average, of every race's average deaths per capita across the U.S.

```python
coviddeaths_percapita = coviddeaths_uscensus.drop(['# of Asian Deaths', '# of Black Deaths', '# of Hispanic Deaths', '# of White Deaths', 'Am. Indian', 'Asian', 'Black', 'Hispanic', 'Total', 'Total Population', 'White'], axis=1)
coviddeaths_percapita.loc['*Average across the U.S.'] = coviddeaths_percapita.mean(axis=0)
coviddeaths_percapita
```

Quickly doing a ANOVA test to look at our results.
Where our Null Hypothesis is that all the average deaths of our Races are all equal, and 
our alternate being that all the average deaths of the four Races are not equal to each other.

```python
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns F and P-value
fvalue, pvalue = stats.f_oneway(coviddeaths_percapita["White deaths per 100K"], coviddeaths_percapita["Black deaths per 100K"], coviddeaths_percapita["Hispanic deaths per 100K"], coviddeaths_percapita["Asian deaths per 100K"])
print("The p-value for this ANOVA test is: " + str(pvalue))
print("The F-statistic for this ANOVA test is: " + str(fvalue) + "\n")

if pvalue < 0.05:
    print("This p-value of " + str(pvalue) + " is smaller than our alpha value of 0.05, \nmeaning we can reject the null hypothesis that all our means are equal. \nIt is very likely that the averages of the deaths by covid19 per capita of our Races are different. ")
else:
    print("This p-value of " + str(pvalue) + " is larger than 5%, meaning we fail to reject the null hypothesis that the averages are all the same.")
```

Here I pull each column of races into a list, so I can more easily use it in my R code. As seen in the next block. I also create a list for the states and the races.

```python
display(coviddeaths_percapita_noavg.dtypes)
covid_colnames = list(coviddeaths_percapita_noavg.columns)
covid_rownames = list(coviddeaths_percapita_noavg.index.values)
#print(covid_rownames)

col_white_list = coviddeaths_percapita_noavg['White deaths per 100K'].tolist()
col_black_list = coviddeaths_percapita_noavg['Black deaths per 100K'].tolist()
col_hisp_list = coviddeaths_percapita_noavg['Hispanic deaths per 100K'].tolist()
col_asian_list = coviddeaths_percapita_noavg['Asian deaths per 100K'].tolist()
col_total_list = coviddeaths_percapita_noavg['Total deaths per 100K'].tolist()
#print(col_asian_list)
```

I use rpy2 here and pull in my variables, lists, and dataframe, so they can be worked in R.
First I used unlist for each column of race values I had, to create vectors in R. Then I created a vector of all the Race names. I then create a dataframe in R using the vectors of Race values. 
I then created 4 barplots of each race to look at each race more closely with their per state deaths per capita. 
Lastly, I took each indiviual vector of race values and averaged them individually. Then i created a vector with all the averages and I made a boxplot from these averages.

```python
%%R -i coviddeaths_percapita_noavg,covid_colnames,covid_rownames,col_white_list,col_black_list,col_hisp_list,col_asian_list,col_total_list

w <- unlist(col_white_list)
b <- unlist(col_black_list)
h <- unlist(col_hisp_list)
a <- unlist(col_asian_list)
races <- factor(c("White","Black","Hispanic","Asian"))
covid_df <- data.frame(white=w,black=b,hispanic=h,asian=a)
states=unlist(covid_rownames)
#print(covid_df)

barplot(w, names = states, ylab = "Deaths Per Capita", main = "White Deaths per Capita Across the U.S.", las = 2, col = "red")
barplot(b, names = states, ylab = "Deaths Per Capita", main = "Black Deaths per Capita Across the U.S.", las = 2, col = "orange")
barplot(h, names = states, ylab = "Deaths Per Capita", main = "Hispanic Deaths per Capita Across the U.S.", las = 2, col = "yellow")
barplot(a, names = states, ylab = "Deaths Per Capita", main = "Asian Deaths per Capita Across the U.S.", las = 2, col = "green")

avg_w <- mean(w)
avg_b <- mean(b)
avg_h <- mean(h)
avg_a <- mean(a)

avgs <- c(avg_w,avg_b,avg_h,avg_a)
barplot(avgs,names=races, ylab = "Deaths per Capita (100K)", main = "Average per Capita Covid19 Deaths in the U.S. per State", xlab = "Races", col = c("red", "orange", "yellow", "green"))
```
